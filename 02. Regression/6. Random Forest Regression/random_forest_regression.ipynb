```python
#@title 1. Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#@title 2. Importing the Dataset

dataset = pd.read_csv("Position_Salaries.csv")

X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1]
```

```python
#@title 3. Training the Random Forest Regression Model on the Whole Dataset
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=10, random_state=0)

regressor.fit(X, y)
```

```python
#@title 4. Predicting A New Result

input_num = 6.5 #@param {type:"number"}

regressor.predict([[input_num]])
```

```python
#@title 5. Visualizing the Random Forest Regression Results (higher resolution)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)

plt.title("Position Salaries")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.scatter(X, y, color="red")
plt.plot(X_grid, regressor.predict(X_grid), color="blue")
```
