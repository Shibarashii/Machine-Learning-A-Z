```python
#@title Importing libraries
!pip install apyori
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#@title Importing the dataset

dataset_name = "Market_Basket_Optimisation.csv" #@param {type:"string"}
dataset = pd.read_csv(dataset_name, header=None)
transactions = []
for i in range(0, len(dataset)):
  transactions.append([str(dataset.values[i, j]) for j in range(0, 20)])
```

```python
#@title Training the Dataset on the Eclat Model
from apyori import apriori

rules = apriori(transactions, min_support=0.003, min_confidence = 0.2, min_lift=3, min_length=2, max_length=2)
```

## Visualizing the Results

```python
#@title Displaying the first results coming directly from the output of the apriori function
results = list(rules)
print(results)
```

```python
#@title Putting the results well organized into a Pandas DataFrame
def inspect(results):
  lhs         = [tuple(result[2][0][0])[0] for result in results]  # Left Hand Side
  rhs         = [tuple(result[2][0][1])[0] for result in results]  # Right Hand Side
  supports    = [result[1] for result in results]

  return list(zip(lhs, rhs, supports))

results_in_data_frame = pd.DataFrame(inspect(results), columns = ["Left Hand Side", "Right Hand Side", "Support"])
```

```python
#@title Displaying the results non-sorted
print(results_in_data_frame)
```

```python
#@title Displaying the results sorted by descending supports
print(results_in_data_frame.nlargest(n=10, columns="Support"))
```
