```python
#@title 1. Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#@title 2. Importing the Dataset

dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
#@title 3. Training the Decision Tree Regression on the Whole Dataset
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

```python
#@title 4. Predicting a New Result
input_number = 6.5 #@param {type: "number"}

regressor.predict([[input_number]])
```

```python
#@title 5. Visualizing the Decision Tree Regression Results (Higher Resolution)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)

plt.title("Position Salaries")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.scatter(X, y, color="red")
plt.plot(X_grid, regressor.predict(X_grid), color="blue")
```
